```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
taxi = pd.read_csv('clean_taxi.csv')
```

```python
taxi.head(10)
```

```python
taxi.describe()
```

```python
taxi = taxi.drop(['lpep_pickup_datetime', 'lpep_dropoff_datetime'],1)
```

```python
taxi.groupby(['payment_type','tip_amount']).mean()
```

```python
taxi.info()
```

```python
le = LabelEncoder()
```

```python
taxi['store_and_fwd_flag'] = le.fit_transform(taxi['store_and_fwd_flag'])
taxi.head()
```

```python
X = taxi.drop('total_amount', axis=1)
y = taxi['total_amount']
```

```python
X = pd.get_dummies(data=X)
y = pd.get_dummies(data=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75, random_state=123)
```

```python
X_train.dtypes
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaler.fit_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import metrics
```

```python
lr = LinearRegression()
dt = DecisionTreeRegressor()
knn = KNeighborsRegressor()
```

```python
lr.fit(X_train, y_train)
lin_pred = lr.predict(X_test)

print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, lin_pred)))
print('R-Squared:', metrics.r2_score(y_test, lin_pred))
```

```python
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)

print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, dt_pred)))
print('R-Squared:', metrics.r2_score(y_test, dt_pred))
```

```python
knn.fit(X_train, y_train)
knn_pred = knn.predict(X_test)

print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, knn_pred)))
print('R-Squared:', metrics.r2_score(y_test, knn_pred))
```

The most accurate model is the Decision Tree Regressor with the RMSE value of ~0.01 and a R2 value of ~0.49
